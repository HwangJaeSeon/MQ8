# Bracket Retention 

```python
# 1.   datetime 
df['date'] = pd.to_datetime(df['date'])

# 2.    
first_visit = df.groupby('fullVisitorId')['date'].min().reset_index()
first_visit.columns = ['fullVisitorId', 'first_visit_date']

# 3.  (    )
df = df.merge(first_visit, on='fullVisitorId', how='left')

# 4.     
df['days_since_first_visit'] = (df['date'] - df['first_visit_date']).dt.days
```

```python
# 5. Bracket Retention  (D+1, D+3, D+7, D+14 )
brackets = [1, 3, 7, 14]
base_users = df[df['days_since_first_visit'] == 0]['fullVisitorId'].nunique()

retention_data = []
for day in brackets:
    retained = df[df['days_since_first_visit'] == day]['fullVisitorId'].nunique()
    rate = retained / base_users
    retention_data.append({'Day': f'D+{day}', 'Retained Users': retained, 'Retention Rate': round(rate, 4)})

bracket_retention_df = pd.DataFrame(retention_data)
print(bracket_retention_df)
```

```python
# Bracket Retention 
plt.figure(figsize=(8, 5))
plt.bar(bracket_retention_df['Day'], bracket_retention_df['Retention Rate'], color='dodgerblue')

#   
for i, rate in enumerate(bracket_retention_df['Retention Rate']):
    plt.text(i, rate + 0.0005, f'{rate:.2%}', ha='center', va='bottom', fontsize=11)

plt.title('Bracket Retention Rate')
plt.xlabel('Day')
plt.ylabel('Retention Rate')
plt.ylim(0, bracket_retention_df['Retention Rate'].max() + 0.005)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# Rolling Retention

```python
# Rolling Retention : D+1, D+3, D+7, D+14     
rolling_days = [1, 3, 7, 14]

#  : D+0   
base_users = df[df['days_since_first_visit'] == 0]['fullVisitorId'].unique()

# Rolling Retention 
rolling_data = []
for day in rolling_days:
    retained_users = df[
        (df['fullVisitorId'].isin(base_users)) &
        (df['days_since_first_visit'] > 0) &
        (df['days_since_first_visit'] <= day)
    ]['fullVisitorId'].nunique()
    
    retention_rate = retained_users / len(base_users)
    rolling_data.append({'Day': f'D+{day}', 'Retained Users': retained_users, 'Rolling Retention Rate': round(retention_rate, 4)})

rolling_retention_df = pd.DataFrame(rolling_data)
rolling_retention_df
```

```python
# Rolling Retention 
plt.figure(figsize=(8, 5))
plt.bar(rolling_retention_df['Day'], rolling_retention_df['Rolling Retention Rate'], color='darkorange')

#   
for i, rate in enumerate(rolling_retention_df['Rolling Retention Rate']):
    plt.text(i, rate + 0.001, f'{rate:.2%}', ha='center', va='bottom', fontsize=11)

plt.title('Rolling Retention Rate')
plt.xlabel('Day')
plt.ylabel('Retention Rate')
plt.ylim(0, rolling_retention_df['Rolling Retention Rate'].max() + 0.01)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#      
alive_set = set(alive_users)
df['retention_group'] = df['fullVisitorId'].isin(alive_set).map({True: 'Alive', False: 'Churned'})

# Cohort    
df['cohort_month'] = df['first_visit_date'].dt.to_period('M')
df['visit_month'] = df['date'].dt.to_period('M')

#  Cohort 
grouped = df.groupby(['retention_group', 'cohort_month', 'visit_month'])['fullVisitorId'].nunique().reset_index()
grouped.columns = ['Group', 'CohortMonth', 'VisitMonth', 'Users']
grouped['MonthsSinceFirstVisit'] = (grouped['VisitMonth'] - grouped['CohortMonth']).apply(lambda x: x.n)

#   
alive_retention = grouped[grouped['Group'] == 'Alive'].pivot(
    index='CohortMonth', columns='MonthsSinceFirstVisit', values='Users'
)
churned_retention = grouped[grouped['Group'] == 'Churned'].pivot(
    index='CohortMonth', columns='MonthsSinceFirstVisit', values='Users'
)

#   
alive_size = alive_retention.iloc[:, 0]
churned_size = churned_retention.iloc[:, 0]

alive_retention_rate = alive_retention.divide(alive_size, axis=0).round(4)
churned_retention_rate = churned_retention.divide(churned_size, axis=0).round(4)
```

```python
# 'productPagesViewed'      
# 0 10      
threshold_range = range(0, 11)
retention_rates = []

for t in threshold_range:
    subset = first_visit_df[first_visit_df['productPagesViewed'] >= t]
    if len(subset) > 0:
        alive_rate = (subset['retention_group'] == 'Alive').mean()
    else:
        alive_rate = 0
    retention_rates.append(alive_rate)

# 
plt.figure(figsize=(8, 5))
plt.plot(threshold_range, retention_rates, marker='o', linestyle='-', color='dodgerblue')
plt.title('Alive Rate by Product Pages Viewed Threshold')
plt.xlabel('Minimum Product Pages Viewed (D+0)')
plt.ylabel('Alive Rate (D+14)')
plt.xticks(threshold_range)
plt.grid(True)
plt.ylim(0, max(retention_rates) + 0.05)
plt.tight_layout()
plt.show()
```

```python
#      : totalPageviews, totalTimeOnSite

def compute_alive_rates(column, max_threshold, step=1):
    thresholds = list(range(0, max_threshold + 1, step))
    rates = []

    for t in thresholds:
        subset = first_visit_df[first_visit_df[column] >= t]
        if len(subset) > 0:
            alive_rate = (subset['retention_group'] == 'Alive').mean()
        else:
            alive_rate = 0
        rates.append(alive_rate)
    
    return thresholds, rates

# 1. totalPageviews
tpv_thresholds, tpv_rates = compute_alive_rates('totalPageviews', 20)

# 2. totalTimeOnSite (: , 30  )
ttos_thresholds, ttos_rates = compute_alive_rates('totalTimeOnSite', 600, step=30)

# 
plt.figure(figsize=(14, 5))

# totalPageviews
plt.subplot(1, 2, 1)
plt.plot(tpv_thresholds, tpv_rates, marker='o', color='teal')
plt.title('Alive Rate by Total Pageviews Threshold')
plt.xlabel('Minimum Total Pageviews (D+0)')
plt.ylabel('Alive Rate (D+14)')
plt.grid(True)
plt.ylim(0, max(tpv_rates) + 0.05)

# totalTimeOnSite
plt.subplot(1, 2, 2)
plt.plot(ttos_thresholds, ttos_rates, marker='o', color='darkorange')
plt.title('Alive Rate by Total Time on Site Threshold')
plt.xlabel('Minimum Time on Site (seconds)')
plt.ylabel('Alive Rate (D+14)')
plt.grid(True)
plt.ylim(0, max(ttos_rates) + 0.05)

plt.tight_layout()
plt.show()
```
